```python
import numpy as np
from persor import BVHparser
from scipy.spatial.transform import Rotation
import matplotlib.pyplot as plt
```

```python
bvhp = BVHparser("./data/iroiro.bvh")
df = bvhp.get_motion_df()
```

```python
fig = plt.figure(figsize=(20, 10))

for joint in [c for c in df.columns if c.endswith("rotation")]:
    if joint == "time":
        continue
    plt.plot(df["time"], df[joint], label=joint)

plt.axvspan(6, 12, color="blue", alpha=0.3)  # 右足を前に出す
plt.axvspan(21, 26, color="green", alpha=0.3)  # 右足を右に出す
plt.axvspan(33, 40, color="yellow", alpha=0.3)  # 右足を右回転する
plt.axvspan(52, 58, color="purple", alpha=0.3)  # 右手を右に出す
plt.axvspan(63, 69, color="red", alpha=0.3)  # 右手を前に出す
plt.axvspan(18, 45, color="gray", alpha=0.1)  # 持ち替え

plt.ylabel("rotation [rad]")
plt.xlabel("time [s]")
plt.xlim([0, 86])
plt.legend()
plt.show()
```
